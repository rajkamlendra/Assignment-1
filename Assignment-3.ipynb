```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri.head(8)
```

```python
fmri.describe().transpose()
```

```python
# checking for information about columns that we have
fmri.info()
```

```python
fmri.nunique()
```

```python
sns.relplot(data = fmri, x = 'timepoint', y = 'signal', hue = 'event', size = 'event', kind ='line')
plt.title('Relation b/w timepoint and signal')
plt.show()
```

```python
sns.relplot(data = fmri, x = 'timepoint', y = 'signal', hue= 'region')
plt.title('Relation b/w timepoint and signal')
plt.show()
```

```python
sns.boxplot(data = fmri, x = 'timepoint', y = 'signal', hue = 'region')
plt.title('Relation b/w timepoint and signal')
plt.show()
```
